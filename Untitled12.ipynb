```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Heart Disease data.csv')
```

```python
df=df.drop_duplicates()
```

```python
sns.heatmap(df.corr(),annot=True)
plt.title('Correlation among all the attributes', fontsize=30)
```

```python
df['target'].value_counts()
```

```python
sns.countplot(x='target' ,data=df,palette='CMRmap' ,width=0.5)
plt.xlabel('Target',fontsize=20)
plt.ylabel('Count', fontsize=20)
plt.title('')
```

```python
sns.countplot(x='sex',hue='target',data=df,width=0.6)
plt.xlabel('Gender',fontsize=25)
plt.ylabel('Count',fontsize=25)
plt.title('Gender Count based on Heart Disease',fontsize=30)
plt.xticks([1,0],['Male','Female'])
plt.show
```

```python
sns.countplot(data=df,x='cp',hue='target',palette='CMRmap')
plt.xlabel('Chest Pain',fontsize=25)
plt.ylabel('Count',fontsize=25)
plt.title('Chest Pain by Targeted People',fontsize=30)
plt.show
```

```python
sns.countplot(x='fbs',hue='target',data=df)
plt.xlabel('Blood Sugar',fontsize=25)
plt.ylabel('Count',fontsize=25)
plt.title('FBS by Targeted People',fontsize=30)
plt.show
```

```python
sns.lineplot(x='cp',y='age',hue='sex',data=df)
plt.xlabel('Chest Pain',fontsize=25)
plt.ylabel('Age',fontsize=25)
plt.title('Chest Pain by Age',fontsize=30)
plt.show
```

```python
sns.histplot(x='trestbps',color='purple',data=df)
plt.xlabel('Trestbps',fontsize=25)
plt.ylabel('Count',fontsize=25)
plt.title('Trestbps',fontsize=30)
```

```python
plt.figure(figsize=(12, 6))  # Increase figure size to provide more space
sns.countplot(x='age', hue='target', data=df)
plt.xlabel('Age', fontsize=25)
plt.ylabel('Count', fontsize=25)
plt.title('Age Count by Heart Disease', fontsize=30)

# Rotate the x-axis labels to avoid overlap
plt.xticks(rotation=45, ha='right')  # Rotate by 45 degrees and align to the right

# Adjust the layout to ensure everything fits well
plt.tight_layout()
```

```python
sns.histplot(x='chol',color='turquoise',data=df)
plt.xlabel('Cholestrol',fontsize=25)
plt.ylabel('Count',fontsize=25)
plt.title('Cholestrol',fontsize=30)
```
